# Graph analysis functions

## MakeGraph()

```python
def MakeGraph(df, thresh):
    """Function to create graph from boot data"""
    
    edge_labels = {} #Dict to store edge labels

    # Visualize the network
    plt.figure(figsize=(75,60))
    
    global ng
    ng = nx.Graph()

    for i in tqdm(df.index):
        if abs(df.loc[i].Mean) > float(thresh):
            node1, node2 = i.split(" <--> ")[0], i.split(" <--> ")[1]
            label = f"{round(df.loc[i].Mean, 2)}  +/- {round(df.loc[i].Std, 2)}"
            weight = df.loc[i].Mean
            if weight > 0:
                ecolor = "green"
            else:
                ecolor="red"

            #Create nodes
            ng.add_nodes_from([node1, node2])
            ng.add_edge(node1, node2, weight=weight, color=ecolor)

            #Create edge labels
            edge_labels[(node1, node2)] = label

    pos = nx.nx_agraph.graphviz_layout(ng) 

    nx.draw_networkx_nodes(ng, pos, node_size=16000, node_color="gold", node_shape="H",\
            linecolor='black', linewidths=1.5, font_size=0.9)\
            .set_edgecolor("black")
    
    colors_list = [ng[s][f]["color"] for s,f in ng.edges]
    
    nx.draw_networkx_edges(ng, pos, width=2, edge_color=colors_list)

    nx.draw_networkx_labels(ng, pos, font_size=12)

    nx.draw_networkx_edge_labels(ng,pos,edge_labels=edge_labels,\
                     font_color='black', font_size=11)
#     plt.show()
```

## Paths()

```python
def Paths(Graph, start, seen=None, path=None):
    """Function to find the longest path in a graph at specified start node"""
    
    if seen is None:
        seen = []
    else:
        seen.append(start)
        
    if path is None: path = [start]


    paths = []
    for i in Graph[start]:
        if i not in seen:
            i_path = path + [i]
            paths.append(tuple(i_path))
            paths.extend(Paths(Graph, i, seen, i_path)) #Recursive implementation of the function
    return paths
```

## LongestPath()

```python
def LongestPath(Graph, start):
    paths = Paths(Graph, start)
    
    pdct = {} #path dictionary; pdct[path] = length of path

    for i in paths:
        pdct[i] = len(i)
        
    max_l = max(pdct.values())
    
    for p,l in pdct.items():

        if l == max_l:
            print("Longest path\n")
            
            for i in range(len(p)):
                try:
                    if Graph[p[i]][p[i-1]]["weight"] > 0:
                        print("    ",round(Graph[p[i]][p[i-1]]["weight"], 2)) #for prettier printing
                    elif Graph[p[i]][p[i-1]]["weight"] < 0:
                        print("   ",round(Graph[p[i]][p[i-1]]["weight"], 2))
                        
                    print("      |")
                    print(p[i])
                    try:
                        if p[i+1]:
                            print("      |")
                    except IndexError:
                        pass
                except KeyError:
                    print(p[i])
                    print("      |")
                    
            print(f"\nLength: {max_l}")
```
