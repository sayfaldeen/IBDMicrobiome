# Create the pre-processing functions to Clean, threshold, and renormalize the data

- Also create a function to check if the data has been properly re-normalized

```python
#Function to Clean noise 
#replace values below the noise-threhsold with a 0

def clean_noise(df, thresh):
    print("starting shape: {}".format(df.shape))
    df1 = df.copy()
    for i in tqdm(df1.columns):
        if df1[i].values.max() < thresh:
            df1.drop(i, axis=1, inplace=True)

    global df2
    df2 = df1.apply(lambda x: x.where(x > thresh, 0), axis=0)


    print("Ending shape: {}\n".format(df2.shape))
    return df2
```

```python
#Function to check re-normalization

def sum_check(df):
    pcnt = 0
    for i in df.index:
        s = df.loc[i].sum()
        if s < 0.99:
            print("{} failed check; sum={}".format(i, s))
        else: 
            pcnt+=1
    print("\n{}/{} samples pass sum_check\n"\
          .format(pcnt, len(df.index)))
```

```python
#Function to threshold at a given prevalence level

def CleanAndThresh(df0, noise_thresh, prev_percent, hist_title): 
    #dataframe, noise-threshold, prevalence-threshold (as percent), hist_title
    
    thresh = float(noise_thresh)
    
    clean_noise(df0, thresh)
    
    prev = (prev_percent/100)
    
    print("\nStarting bacteria prior to {}% prevalence-threshold: {}".format(prev_percent, len(df0.columns)))
    
    global df
    #global dft
    
    dft = df2.T.copy()

    min_t = int(len(df2)* prev) #threshold point (minimum number of samples to meet the threshold)

    dft["AboveNoise"] = dft.apply(lambda x: x[x > 0].count(), axis=1).values
    
    df_non = dft[dft.AboveNoise >= min_t].T.drop("AboveNoise").copy() #non-renormalized df
    
    df = df_non.apply(lambda x:x/x.sum(), axis=1)
    
    df.dropna(inplace=True, thresh=10)
    df.fillna(0, inplace=True)
    
    print("Ending bacteria: {}".format(len(df.columns)))
    
    sum_check(df)
    
    plt.figure(figsize=(10,8))
    dft.AboveNoise.hist()
    
    plt.title("{} @ {} noise-threshold and {}% prevalence \n"\
              .format(hist_title, thresh, prev_percent), weight='bold')
    
    plt.xlabel("\nNumber of Samples Bacteria are Present in")
    plt.ylabel("Number of Bacteria\n")
    
    plt.axvline(min_t, color='maroon', linestyle='dashed')
    plt.annotate("{}% Threshold"\
                 .format(prev_percent), xy=((min_t-(min_t/30)), len(dft.index)/3), rotation=90, color='maroon')
    plt.annotate("n = {}".format(len(df.columns)), xy=((min_t+ (min_t/30)), (len(df.columns)+ (len(df.columns)/20))),\
                                                      color='maroon', weight='bold')
    

    #plt.show()
```
